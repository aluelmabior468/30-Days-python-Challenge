```python
# 3. Sum arbitrary number of arguments
def add_all_nums(*args):
    total = 0
    for num in args:
        if not isinstance(num, (int, float)):
            return f"Error: {num} is not a number"
        total += num
    return total

print("Sum of 1, 2, 3:", add_all_nums(1, 2, 3))
print("Sum of 1, 'a', 3:", add_all_nums(1, 'a', 3))
```

```python
# 5. Check season
def check_season(month):
    month = month.capitalize()
    if month in ["September", "October", "November"]:
        return "Autumn"
    elif month in ["December", "January", "February"]:
        return "Winter"
    elif month in ["March", "April", "May"]:
        return "Spring"
    elif month in ["June", "July", "August"]:
        return "Summer"
    else:
        return "Invalid month"

print("Season in March:", check_season("March"))
```

```python
# 6. Calculate slope of a line given two points
def calculate_slope(x1, y1, x2, y2):
    if x2 - x1 == 0:
        return "Slope is undefined"
    slope = (y2 - y1) / (x2 - x1)
    return slope

print("Slope between (2,2) and (6,10):", calculate_slope(2, 2, 6, 10))
```

```python
# 7. Solve quadratic equation ax^2 + bx + c = 0
def solve_quadratic_eqn(a, b, c):
    discriminant = cmath.sqrt(b**2 - 4*a*c)
    x1 = (-b + discriminant)/(2*a)
    x2 = (-b - discriminant)/(2*a)
    return (x1, x2)

print("Solutions of x^2 + 5x + 6 = 0:", solve_quadratic_eqn(1, 5, 6))
```

```python
#day 12
import random
import string

# 1. Generate a 6-character random user ID
def random_user_id():
    chars = string.ascii_letters + string.digits
    return ''.join(random.choice(chars) for _ in range(6))

print("Random user ID:", random_user_id())
```

```python
# 3. Generate random RGB color
def rgb_color_gen():
    r = random.randint(0, 255)
    g = random.randint(0, 255)
    b = random.randint(0, 255)
    return f"rgb({r},{g},{b})"

print("Random RGB color:", rgb_color_gen())
```

```python
import requests
import statistics

cats_api = 'https://api.thecatapi.com/v1/breeds'
response = requests.get(cats_api)
cats_data = response.json()

# Extract weight (metric) and lifespan
weights = []
lifespans = []
countries = []
breeds = []

for cat in cats_data:
    # Weight format: "3 - 5"
    w_min, w_max = map(int, cat['weight']['metric'].split(' - '))
    weights.append((w_min + w_max) / 2)  # average weight
    l_min, l_max = map(int, cat['life_span'].split(' - '))
    lifespans.append((l_min + l_max) / 2)
    countries.append(cat.get('origin', 'Unknown'))
    breeds.append(cat.get('name', 'Unknown'))

# Statistics
def stats(data):
    return {
        'min': min(data),
        'max': max(data),
        'mean': round(statistics.mean(data), 2),
        'median': statistics.median(data),
        'std_dev': round(statistics.stdev(data), 2)
    }

print("Cats weight stats (kg):", stats(weights))
print("Cats lifespan stats (years):", stats(lifespans))

# Frequency tables
from collections import Counter
print("Country frequency table:", Counter(countries))
print("Breed frequency table:", Counter(breeds))
```

```python
#day 21
from collections import Counter
import math

class Statistics:
    def __init__(self, data):
        self.data = data
        self.data.sort()
    
    def count(self):
        return len(self.data)
    
    def sum(self):
        return sum(self.data)
    
    def min(self):
        return min(self.data)
    
    def max(self):
        return max(self.data)
    
    def range(self):
        return self.max() - self.min()
    
    def mean(self):
        return round(sum(self.data)/len(self.data), 1)
    
    def median(self):
        n = len(self.data)
        mid = n // 2
        if n % 2 == 0:
            return (self.data[mid-1] + self.data[mid]) / 2
        return self.data[mid]
    
    def mode(self):
        count = Counter(self.data)
        max_count = max(count.values())
        mode = [k for k, v in count.items() if v == max_count]
        return (mode[0], max_count)  # returning first mode
    
    def var(self):
        mean = self.mean()
        return round(sum((x - mean)**2 for x in self.data) / len(self.data), 1)
    
    def std(self):
        return round(math.sqrt(self.var()), 1)
    
    def freq_dist(self):
        count = Counter(self.data)
        total = len(self.data)
        freq = [(round(v/total*100, 1), k) for k, v in count.items()]
        freq.sort(reverse=True)
        return freq
    
    def describe(self):
        return f"""Count: {self.count()}
Sum: {self.sum()}
Min: {self.min()}
Max: {self.max()}
Range: {self.range()}
Mean: {self.mean()}
Median: {self.median()}
Mode: {self.mode()}
Variance: {self.var()}
Standard Deviation: {self.std()}
Frequency Distribution: {self.freq_dist()}"""

# Example usage
ages = [31, 26, 34, 37, 27, 26, 32, 32, 26, 27, 27, 24, 32, 33, 27, 25, 26, 38, 37, 31, 34, 24, 33, 29, 26]
data = Statistics(ages)
print(data.describe())
```

```python
#DAY 22
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json

def save_to_json(data, filename):
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=2)

#Task A: Scrape BU facts & stats page and store as JSON ---
def scrape_bu_stats():
    url = 'http://www.bu.edu/president/boston-university-facts-stats/'
    resp = requests.get(url)
    if resp.status_code != 200:
        print("Failed to fetch BU page:", resp.status_code)
        return
    soup = BeautifulSoup(resp.content, 'html.parser')
    # You need to inspect the page to know where the data is.
    # Example — maybe there are <table> elements, or <div> lists; adapt accordingly.
    data = {}
    # Suppose there's a table:
    table = soup.find('table')
    if table:
        df = pd.read_html(str(table))[0]
        data = df.to_dict(orient='records')
    else:
        # fallback: maybe stats are in <p> or <li>; you can extract manually
        data['text'] = soup.get_text(separator='\n')
    save_to_json(data, 'bu_stats.json')
    print("Saved BU stats to bu_stats.json")
```

```python

# Task B: Extract table from UCI datasets page and store as JSON ---
def scrape_uci_datasets():
    url = 'https://archive.ics.uci.edu/ml/datasets.php'
    resp = requests.get(url)
    if resp.status_code != 200:
        print("Failed to fetch UCI page:", resp.status_code)
        return
    soup = BeautifulSoup(resp.content, 'html.parser')
    # UCI page tables: many — but pandas.read_html can fetch them all
    tables = pd.read_html(resp.text)
    all_data = []
    for df in tables:
        # convert each table to a list of dicts
        records = df.to_dict(orient='records')
        all_data.append(records)
    save_to_json(all_data, 'uci_datasets.json')
    print("Saved UCI datasets to uci_datasets.json")
```

```python
# --- Task C: Scrape US Presidents list from Wikipedia and store as JSON ---
def scrape_us_presidents():
    url = 'https://en.wikipedia.org/wiki/List_of_presidents_of_the_United_States'
    resp = requests.get(url)
    if resp.status_code != 200:
        print("Failed to fetch Wikipedia page:", resp.status_code)
        return
    soup = BeautifulSoup(resp.content, 'html.parser')
    # Find the main presidents table — typically class 'wikitable'
    table = soup.find('table', class_='wikitable')
    if not table:
        print("Could not find a wikitable on the page.")
        return
    df = pd.read_html(str(table))[0]
    # Convert to JSON
    records = df.to_dict(orient='records')
    save_to_json(records, 'us_presidents.json')
    print("Saved US presidents data to us_presidents.json")

if __name__ == "__main__":
    scrape_bu_stats()
    scrape_uci_datasets()
    scrape_us_presidents()
```

#day 28 - API
API
API stands for Application Programming Interface. The kind of API we will cover in this section is going to be Web APIs. Web APIs are the defined interfaces through which interactions happen between an enterprise and applications that use its assets, which also is a Service Level Agreement (SLA) to specify the functional provider and expose the service path or URL for its API users.

In the context of web development, an API is defined as a set of specifications, such as Hypertext Transfer Protocol (HTTP) request messages, along with a definition of the structure of response messages, usually in an XML or a JavaScript Object Notation (JSON) format.

Web API has been moving away from Simple Object Access Protocol (SOAP) based web services and service-oriented architecture (SOA) towards more direct representational state transfer (REST) style web resources.

Social media services, web APIs have allowed web communities to share content and data between communities and different platforms.

Using API, content that is created in one place dynamically can be posted and updated to multiple locations on the web.

For example, Twitter's REST API allows developers to access core Twitter data and the Search API provides methods for developers to interact with Twitter Search and trends data.
